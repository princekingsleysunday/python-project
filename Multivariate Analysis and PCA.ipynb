```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('USA_cars_datasets (1).csv')
```

```python
data
```

```python
data.drop('Unnamed: 0', axis =1, inplace = True)
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
```

```python
data['color'] = encoder.fit_transform(data['color'])
data['brand'] = encoder.fit_transform(data['brand'])
data['model'] = encoder.fit_transform(data['model'])
```

```python
# extract the features(independednt variable) and labels (dependent variable)

# value is a keyword to convert the df to an array
x = data.loc[:,['lot','brand','color','model','year']].values
y= data.iloc[:,0].values
```

# PCA

```python
# step 1: standardizing the features
# assuming the dataset follows a normal distribution

from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_scaled = sc.fit_transform(x)
```

```python
x_scaled
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 4)

p = pca.fit_transform(x_scaled)
```

```python
p
```

```python
principalDF = pd.DataFrame(data=p, columns= ['PC1', 'PC2', 'PC3', 'PC4'])
finalDf = pd.concat([principalDF, data[['price']]], axis = 1)
finalDf.head()
```
